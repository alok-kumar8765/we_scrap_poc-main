# Bar Graph

```python
# pivot_table function to find the total number of first dose for each states:
table = pd.pivot_table(data=df,index='State',values='Total Individuals Vaccinated')
table
```

```python
plt.figure(figsize=(15,4))
#my_dpi=120
#plt.figure(figsize=(400/my_dpi, 300/my_dpi), dpi=my_dpi)
#bar graph
plt.bar(table.index,table['Total Individuals Vaccinated'])

#xticks 
plt.xticks(rotation=90) 

#x-axis labels 
plt.xlabel('State') 

#y-axis labels 
plt.ylabel('Total Vaccinated') 

#plot title 
plt.title('Vaccinated In States') 

#save plot 
plt.savefig('matplotlib_plotting_6.png',dpi=300,bbox_inches='tight') 

#display 
plt.show()
```

# scatter plot

```python
state = df_scrap["State"]
total = df_scrap["Total Individuals Vaccinated"]
x=[]
y=[]
x=list(state)
y=list(total)
plt.xticks(rotation=90)
plt.scatter(x,y)
plt.xlabel('STATE')
plt.ylabel('Total Vaccinated->')
plt.title('Data')
plt.show()
```

```python
#for each pie wedge, the format string is '%. 2f'
#numerical percent value for that wedge is pct
plt.pie(y,labels=x,autopct='%.2f%%')
```

```python
state = df_scrap["State"]
total = df_scrap["Total Individuals Vaccinated"]
x=[]
y=[]
x=list(state)
y=list(total)
my_colors = ['lightblue','red','silver','pink','white','blue','green','black','yellow']
my_explode = (0, 0.1, 0,0,0,0,0.3,0,0.1)
#autopctenables you to display the percent value using Python string formatting
plt.pie(y,labels=x,autopct='%1.1f%%',colors=my_colors,startangle=60,shadow=True,explode=my_explode)
plt.title('Vaccine Pie')
plt.axis('equal')
plt.show()
```
