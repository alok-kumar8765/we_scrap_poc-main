# With Pandas Fetch Tabels From WHO Web Site

```python
#Step1:- Import pandas and assign web site url
```

```python
import pandas as pd
url = 'https://covid19.who.int/info/'
df_list = pd.read_html(url)
#df means datafram, it help to fram table from web page to out terminal output
```

```python
df_list
```

```python
# we can check how many tables are present in this page
len(df_list)
```

# Head()

```python
df_list[0].head()
#here we are going to the very first table in page and .head() helps to show first 5 elements in row
```

```python
#if you need you can print more then or less then 5 rows from top
df_list[0].head(7)
```

```python
#with help of skiprows you can skip single or multiple row and print new datafram
#2 ke pahle ke rows show nhi honge aur, 2 kud ek table header show krega
df=pd.read_html(url,skiprows=2)[0]
df
```

```python
#multiple skip
df=pd.read_html(url,skiprows=[2,4,6])[0]
df
```

```python
#here we first skip 2 rows then header goes to 3rd place after skip and make those rows as header of table
df=pd.read_html(url,skiprows=2,header=3)[0]
df
```

# Drop & Re-Indexing

```python
df = df_list[0].drop([0, 1,3, 7], axis=0)
df
#if you want to eliminate any row simply use .drop([row_1,row_3]) and axis=0 means we are eliminating rows
```

```python
df = df.reset_index(drop=False)
df
#drop=Flase show our after and before index
```

```python
df = df.reset_index(drop=True)
df
#this reset_index() help to reset the elememt index
#If we don’t add the drop=True parameter, the current index will become a new column.
```
