```python
import urllib.request
# Open the specific URL for reading using urlopen()
request_url = urllib.request.urlopen('https://covid19.who.int/info/')
# Print the response data of the URL
print("The output of the URL is:\n\n",request_url.read())
#string return of web page html code
```

# Parsing and unparsing URLs with urllib.parse

```python
# Import parse module of urllib
#helps to define functions to manipulate URLs and their components parts, to build or break them.
import urllib.parse
# Parsing URL using urlparse()

urlParse = urllib.parse.urlparse('https://covid19.who.int/info/')
#splitting a URL into small components
print("\nThe output of URL after parsing:\n", urlParse)

# Joining URL using urlunparse()
urlUnparse = urllib.parse.urlunparse(urlParse)
#joining different URL components into URL string
print("\nThe joining output of parsing URL:\n", urlUnparse)
```

```python
# Parsing URL using urlsplit()
urlSplit = urllib.parse.urlsplit(' https://covid19.who.int/info/')
#split every segment and give only netloc detail
print('\nThe output of URL after splitting:\n', urlSplit)


# Joining URL using urlunsplit()

urlUnsplit = urllib.parse.urlunsplit(urlSplit)

print('\nThe joining output of splitting URL:\n',urlUnsplit)
```

# Reading response header of HTML with urllib.request

```python
urlResponse = urllib.request.urlopen('https://covid19.who.int/info/')

# Reading info like date,Last modify,server etc
print(urlResponse.info())
```

```python
# Reading header information separately
#server using
print('Response server = ', urlResponse.info()["Server"])
#current date time
print('Response date is = ', urlResponse.info()["Date"])
#html/text
print('Response content type is = ', urlResponse.info()["Content-Type"])

```

# Reading URL responses line by line

```python
response = urllib.request.urlopen('https://covid19.who.int/info/')

# Read the URL from response
print ('URL:', response.geturl())

# Read the response text line by line
print('\nReading content:')

for line in response:
    print(line.strip())

#The strip() method is used to remove the space from both sides of each line
```

```python
import urllib.error
# try block to open any URL for reading

try:
   url = input("Enter any URL address: ")
   response = urllib.request.urlopen(url)
   print(response.read())

# Catch the URL error that will generate when opening any URL
except urllib.error.URLError as e:
   print("URL Error:",e.reason)


# Catch the invalid URL error
except ValueError:
  print("Enter a valid URL address")

```
